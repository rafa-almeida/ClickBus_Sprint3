```python
from datetime import datetime, timedelta
```

```python
def verificar_premium(df, id_cliente):
    # Verifica se um cliente atinge o status premium com base no histórico de compras no DataFrame.
    # Args:
    #    df (pd.DataFrame): O DataFrame contendo os dados de compra.
    #    id_cliente (str): O ID do cliente a ser verificado (coluna 'fk_contact').

    # Returns:
    #     bool: True se o cliente é premium, False caso contrário.
    agora = datetime.now()
    um_ano_atras = agora - timedelta(days=365)

    # Verificações através das datas de compras dos clientes
    df['date_purchase'] = pd.to_datetime(df['date_purchase'])

    compras_cliente = df[(df['fk_contact'] == id_cliente) &
                         (df['date_purchase'] >= um_ano_atras) &
                         (df['date_purchase'] <= agora)]

    total_passagens = compras_cliente['total_tickets_quantity_success'].sum()
    return total_passagens >=6
```

```python
# Condição que define cliente comum e cliente Premia.

# Use um ID de cliente de exemplo do seu DataFrame
if not df.empty:
    id_verificar = df['fk_contact'].iloc[2] # é realizada a verificação através da coluna fk_contact
else:
    id_verificar = "ID_DE_CLIENTE_DE_EXEMPLO" # Substitua por um ID real se o DataFrame estiver vazio

if verificar_premium(df, id_verificar):
    print(f"O cliente {id_verificar} é Premium.")
else:
    print(f"O cliente {id_verificar} não é Premium.")
```
